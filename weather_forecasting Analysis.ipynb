```python
import requests
import json
caps=['Hyderabad',
 'Patna',
 'Agartala',
 'Kohima',
 'Chandigarh',
 'Ranchi',
 'Thiruvananthapuram',
 'Mumbai',
 'Bhubaneswar',
 'Chandigarh',
 'Panaji',
 'Gandhinagar',
 'Dispur',
 'Amaravati',
 'Dehradun',
 'Shimla',
 'Bhopal',
 'Jaipur',
 'Aizawl',
 'Srinagar',
 'Kolkata',
 'Raipur',
 'Chennai',
 'Lucknow',
 'Gangtok',
 'Itanagar',
 'Bangalore',
 'Shillong',
 'Imphal']
temp = []
weather = []
city = []
for i in caps:
    url=requests.get("enter your api here") #please get your api from openweather.com
    data=json.loads(url.text)
    weather.append(data['weather'][0]['main'])
    temp.append(data['main']['temp'])
    city.append(data["name"])
```

```python
import numpy as np
import matplotlib.pyplot as plt
x=np.linspace(0,3*np.pi,101)
y=np.sin(x)
z=np.sinh(x)
fig,ax1=plt.subplots()#separate the figure object and axes object from the plotting object
#duplicate the axes with different y-axes and same x-axes
ax2=ax1.twinx()
ax1.plot(x,y,label='sin x',color="r")
ax2.plot(x,z,label='sinh',color='g')
plt.title("plot of sin and hyperbolic sin")
plt.legend()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
x=np.linspace(0,3*np.pi,101)
y=np.sin(x)
z=np.sinh(x)
fig,ax1=plt.subplots()#separate the figure object and axes object from the plotting object
#duplicate the axes with different y-axes and same x-axes
ax2=ax1.twinx()
curve1,=ax1.plot(x,y,label='sin x',color="r")
curve2,=ax2.plot(x,z,label='sinh',color='g')
curves=[curve1,curve2]
#ax1.legend(curves,[curve.get_label() for curve in curves])
ax2.legend(curves,[curve.get_label() for curve in curves])
ax1.set_xlabel("Angle/value",color=curve1.get_color())#in this color value is extracted from curve1 i.e;  red color
ax2.set_ylabel("Magnitude",color=curve1.get_color())
#ax2.set_ylabel("Magnitude",color=curve1.get_color())#ax2 has no property control over x-axes w.r.t colr and grid and label
ax1.set_ylabel("Magnitude",color=curve1.get_color())
plt.title("plot of sin and hyperbolic sin")
plt.show()
plt.savefig("mat.jpg")
```
